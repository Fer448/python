```python
import pandas as pd
import prince
import matplotlib.pyplot as plt

# Datos proporcionados
data = {
    'Distrito': ['131. Entrevas', '132. San Diego', '133. Palomeras Bajas',
                 '134. Palomeras Sureste', '135. Portazgo', '136. Numancia'],
    'PP': [3079, 3486, 4656, 5159, 3286, 5623],
    'MM-VQ': [3141, 3340, 5525, 5136, 2820, 4510],
    'P.S.O.E.': [4236, 3485, 4577, 5287, 3154, 4373],
    'VOX': [1147, 1037, 1263, 1594, 942, 1471],
    'PODEMOS-IU-AV': [1106, 1117, 2175, 1675, 1018, 1400]
}

# Crear DataFrame
df = pd.DataFrame(data)
df.set_index('Distrito', inplace=True)

# Realizar el anlisis de correspondencias simples
ca = prince.CA(n_components=2)  # Reducir a 2 dimensiones
ca.fit(df)

# Obtener las coordenadas de las filas (distritos) y columnas (partidos)
row_coords = ca.row_coordinates(df)
col_coords = ca.column_coordinates(df)

# Visualizar los resultados
plt.figure(figsize=(10, 8))
plt.scatter(row_coords[0], row_coords[1], label='Distritos', color='blue')
plt.scatter(col_coords[0], col_coords[1], label='Partidos', color='red')

# Etiquetar puntos
for i, txt in enumerate(df.index):
    plt.annotate(txt, (row_coords[0][i], row_coords[1][i]), fontsize=9, color='blue')
for i, txt in enumerate(df.columns):
    plt.annotate(txt, (col_coords[0][i], col_coords[1][i]), fontsize=9, color='red')

# Lneas de referencia
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)

# Etiquetas y ttulo
plt.xlabel('Dimensin 1')
plt.ylabel('Dimensin 2')
plt.title('Anlisis de Correspondencias Simples (ACS)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
print("Coordenadas de los distritos:")
print(row_coords)

print("\nCoordenadas de los partidos:")
print(col_coords)
```
