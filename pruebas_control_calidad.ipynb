```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Datos
data = {
    "Atributo_1": [3, 2, 4, 5, 2, 3, 4, 2, 3, 4, 1, 4, 2, 3, 4],
    "Atributo_2": [3, 3, 3, 1, 2, 3, 3, 3, 2, 2, 2, 5, 2, 3, 2],
    "Atributo_3": [3, 4, 3, 5, 2, 3, 4, 2, 4, 4, 1, 4, 3, 4, 4],
    "Atributo_4": [2, 3, 2, 1, 2, 3, 3, 3, 2, 2, 2, 3, 3, 4, 2],
    "Atributo_5": [4, 2, 4, 5, 2, 3, 4, 2, 3, 4, 1, 2, 2, 4, 2]
}

# Convertir a DataFrame
df = pd.DataFrame(data)

# Calcular la media de cada atributo
medias = df.mean()

# Gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x=medias.index, y=medias.values, palette="viridis")
plt.title("Media de Calidad Percibida por Atributo", fontsize=16)
plt.xlabel("Atributos", fontsize=14)
plt.ylabel("Media de Puntuación", fontsize=14)
plt.ylim(0, 5)  # Límite del eje Y para que coincida con la escala de puntuación
plt.show()
```

```python
import numpy as np

# Función para calcular límites de control
def calcular_limites_control(datos):
    media = np.mean(datos)
    desviacion = np.std(datos)
    limite_superior = media + 3 * desviacion
    limite_inferior = media - 3 * desviacion
    return limite_superior, limite_inferior

# Gráficos de control por atributo
plt.figure(figsize=(14, 10))
for i, atributo in enumerate(df.columns, 1):
    plt.subplot(3, 2, i)
    plt.plot(df[atributo], marker="o", linestyle="-", color="b", label=atributo)

    # Calcular límites de control
    limite_superior, limite_inferior = calcular_limites_control(df[atributo])

    # Líneas de control
    plt.axhline(limite_superior, color="r", linestyle="--", label="Límite Superior")
    plt.axhline(limite_inferior, color="r", linestyle="--", label="Límite Inferior")
    plt.axhline(np.mean(df[atributo]), color="g", linestyle="-", label="Media")

    plt.title(f"Gráfico de Control - {atributo}", fontsize=12)
    plt.xlabel("Muestras", fontsize=10)
    plt.ylabel("Puntuación", fontsize=10)
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# Definir la matriz de correlación
matriz_correlacion = np.array([
    [1.00, 0.15, 0.45, -0.10, 0.30],
    [0.15, 1.00, 0.20, 0.05, -0.25],
    [0.45, 0.20, 1.00, 0.30, 0.50],
    [-0.10, 0.05, 0.30, 1.00, 0.10],
    [0.30, -0.25, 0.50, 0.10, 1.00]
])

# Nombres de los atributos
atributos = ['Atributo_1', 'Atributo_2', 'Atributo_3', 'Atributo_4', 'Atributo_5']
# Configurar el tamaño de la figura
plt.figure(figsize=(8, 6))

# Crear el mapa de calor
sns.heatmap(
    matriz_correlacion,
    annot=True,  # Mostrar los valores en las celdas
    fmt=".2f",   # Formato de los números (2 decimales)
    cmap="coolwarm",  # Paleta de colores
    xticklabels=atributos,  # Etiquetas del eje X
    yticklabels=atributos   # Etiquetas del eje Y
)

# Añadir título
plt.title("Mapa de Calor de la Matriz de Correlación", fontsize=16)

# Mostrar el gráfico
plt.show()
```
